```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

### Discrete distributions

```python
rv_bernoulli = stats.bernoulli(p=0.3) # Bernoulli distribution
```

```python
rv_bernoulli.rvs(14)
```

```python
rv_binom = stats.binom(100, p=0.9) # Binomial distribution (the Bernoulli distribution is a special case of the binomial distribution) 
```

```python
rv_binom.rvs(8)
```

```python
plt.hist(rv_binom.rvs(80), bins=10)
```

### Continuous distributions

```python
a = 5
b = 10
rv_unifrom = stats.uniform(a, b-a) # Uniform distribution
```

```python
rv_unifrom.cdf(5.5) # Cumulative distribution function
```

```python
rv_unifrom.pdf(7) # Probability distribution function
```

```python
X = np.linspace(a - 2, b + 2, 100)
cdf = rv_unifrom.cdf(X)
plt.plot(X, cdf)

plt.xlabel('x')
plt.ylabel('F(x)')
plt.ylim(0, 1.5)
plt.title(u'Cummulative distribution function for uniform')
```

```python
X = np.linspace(a - 2, b + 2, 1000)
pdf = rv_unifrom.pdf(X)
plt.plot(X, pdf)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([4, 12])
plt.title(u'Probability distribution function for uniform')
```

```python
mu = 2
sigma = 0.5
rv_norm = stats.norm(loc=mu, scale=sigma) # Normal distribution
```

```python
rv_norm.rvs(17)
```

```python
x = np.linspace(0, 4, 100)
cdf = rv_norm.cdf(x)
plt.plot(x, cdf)

plt.xlabel('x')
plt.ylabel('F(x)')
plt.title(u'Cummulative distribution function for normal')
```

```python
x = np.linspace(0, 4, 100)
pdf = rv_norm.pdf(x)
plt.plot(x, pdf)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(u'Probability distribution function for normal')
```

```python
means = [1, 5, 10]
mu = 10
sigmas = [1, 2, 4, 6]
for sigma in sigmas:
    rv_norm = stats.norm(loc=mu, scale=sigma)
    x = np.linspace(0, 20, 200)
    cdf = rv_norm.pdf(x)
    plt.plot(x, cdf, label=sigma)
plt.legend()
```
