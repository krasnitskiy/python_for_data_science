### Столбчатая диаграмма среднего уровня дохода игроков некоторых футбольных клубов

Столбчатая диаграмма позволяет сравнить между собой некоторые (наибролее часто упоминаемые в датасете) футбольные клубы по размеру среднего заработка своих футболистов (что косвенно показывает финансовую состоятельность клубов).

```python
# импортируем необходимые библиотеки
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fp = pd.read_csv('CompleteDataset.csv', low_memory=False) # читаем датасет (создаем датафрейм)
```

```python
fp.head() # проверяем, что все корректно прочиталось
```

```python
fp.Wage = fp.Wage.str.strip('€').str.replace('K', '000') # удаляем лишние знаки из столбца дохода 
```

```python
fp.Wage = fp.Wage.astype(int) # приводим столбец дохода к числовому типу
```

```python
wage_club = fp.groupby(['Club'])[['Wage']].mean() # расчитываем средний доход игроков, группируя их по-клубно 
```

```python
wage_club # проверяем, что все сделали корректно
```

```python
fp['Club'].value_counts() # перечень клубов по убыванию числу упоминаний игроков в датасете
```

```python
# выводим доход клубов с наибольшим числом игроков в датасете (34 и 35)
wage_club.loc[['VfL Wolfsburg', 
               'VfL Wolfsburg', 
               'UD Las Palmas', 
               'FC Nantes', 
               'Borussia Dortmund', 
               'Manchester United', 
               'OGC Nice', 
               'Olympique Lyonnais']] 
```

```python
# Строим столбчатую диаграмму
ax = wage_club.loc[['VfL Wolfsburg', 
               'VfL Wolfsburg', 
               'UD Las Palmas', 
               'FC Nantes', 
               'Borussia Dortmund', 
               'Manchester United', 
               'OGC Nice', 
               'Olympique Lyonnais']].plot.bar(color='green') 
plt.title('Average wage of players')
plt.xlabel('Club')
plt.ylabel('Average wage')
```
