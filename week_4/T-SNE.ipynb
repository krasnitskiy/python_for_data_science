```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
from sklearn.manifold import TSNE
```

```python
from sklearn.datasets import load_digits
```

```python
digit = load_digits()
```

```python
img = digit.images
```

```python
img.shape
```

```python
plt.imshow(img[0], cmap=plt.cm.Greys_r)
```

```python
X = img.reshape(-1, 64)
```

```python
X.shape
```

```python
tnse = TSNE(n_components=2, perplexity=30.)
```

```python
tnse.fit(X)
```

```python
Z = tnse.embedding_
```

```python
Z.shape
```

```python
y = digit.target
```

```python
plt.scatter(Z[:, 0], Z[:, 1], c=y)
```
